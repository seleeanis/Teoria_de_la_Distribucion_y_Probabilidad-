## Definición del problema

Se modela una variable aleatoria X que representa el número de éxitos en 10 intentos,
donde la probabilidad de éxito en cada intento es 0.5.

Esta variable sigue una distribución binomial:
X ~ Binomial(n=10, p=0.5)

```python
# Importación de librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
import seaborn as sns
import itertools

# Configuración visual
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12,5)
```

```python
# Parámetros de la distribución binomial
n = 10
p = 0.5

# Valores posibles
x = np.arange(0, n+1)

# PMF y CDF
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)
```

```python
plt.figure(figsize=(12,5))

# PMF
plt.subplot(1,2,1)
plt.stem(x, pmf)
plt.title("PMF - Distribución Binomial")
plt.xlabel("Número de éxitos")
plt.ylabel("P(X = x)")

# CDF
plt.subplot(1,2,2)
plt.step(x, cdf, where='post')
plt.title("CDF - Distribución Binomial")
plt.xlabel("Número de éxitos")
plt.ylabel("P(X ≤ x)")

plt.show()

```

```python
# Probabilidad de exactamente 5 éxitos
p1 = binom.pmf(5, n, p)

# Probabilidad de máximo 3 éxitos
p2 = binom.cdf(3, n, p)

# Probabilidad entre 4 y 7 éxitos
p3 = binom.cdf(7, n, p) - binom.cdf(3, n, p)

print("P(X = 5):", p1)
print("P(X ≤ 3):", p2)
print("P(4 ≤ X ≤ 7):", p3)
```

```python
# Distribución normal (continua)
mu = 200
sigma = 30

x_cont = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x_cont, mu, sigma)

plt.plot(x_cont, pdf)
plt.title("PDF - Distribución Normal")
plt.xlabel("Tiempo")
plt.ylabel("f(x)")
plt.show()
```

```python
# Generación de espacio muestral con itertools
espacio = list(itertools.product(['H', 'T'], repeat=2))
print("Espacio muestral:", espacio)
```

## Interpretación de resultados

En la distribución binomial con n = 10 y p = 0.5:

- P(X = 5): Representa la probabilidad de obtener exactamente 5 éxitos en 10 intentos.
Se observa que este valor es relativamente alto, lo cual tiene sentido porque 5 es el valor esperado cuando la probabilidad de éxito es 0.5.

- P(X ≤ 3): Representa la probabilidad de obtener como máximo 3 éxitos.
Este valor es menor, indicando que obtener pocos éxitos es menos probable en este escenario.

- P(4 ≤ X ≤ 7): Representa la probabilidad de obtener un número intermedio de éxitos.
Este rango concentra una mayor probabilidad, ya que los valores cercanos al centro de la distribución son más probables.

En la distribución normal:

La gráfica de la PDF muestra que los valores cercanos a la media (μ = 200) tienen mayor densidad, lo que indica que los tiempos de respuesta del servidor tienden a concentrarse alrededor de ese valor.

A diferencia del caso discreto, en la distribución continua la probabilidad no se mide en puntos específicos, sino como el área bajo la curva en un intervalo.
